## Using Mata

```stata
sysuse auto, clear
```

```stata
mata
```

```stata
%status
```

```stata
y    = st_data(., "price")
X    = st_data(., "mpg trunk")
n    = rows(X)
X    = X,J(n,1,1)
XpX  = quadcross(X, X)
XpXi = invsym(XpX)
b    = XpXi*quadcross(X, y)
```

```stata
b'
```
